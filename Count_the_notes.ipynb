```python
amount = int(input("enter the number of amount:"))

note_1 = amount//100
note_2 = (amount%100)//50
note_3 = ((amount%100)%50)//10

print("notes of 100 is ", note_1)
print("notes of 50 is ", note_2)
print("notes of 10 is ", note_3)
```
